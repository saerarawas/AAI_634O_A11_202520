```python
import pandas as pd
from pymongo import MongoClient
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
import plotly.express as px

# MongoDB connection setup
connection_string = "mongodb+srv://tsjannoun123:KufyyNNqnno0atX9@cluster0.sb8py.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"
client = MongoClient(connection_string)
db = client['weather']
sales = db['sales']

# Query data from MongoDB
sales_data = pd.DataFrame(list(sales.find()))

# Preprocess the data (e.g., convert date to datetime type)
sales_data['date'] = pd.to_datetime(sales_data['date'])

# Check the first few rows of the data
print(sales_data.head())
```

```python
# Initialize the Dash app
app = dash.Dash(__name__)

# App layout with a title and a graph for visualization
app.layout = html.Div(children=[
    html.H1("Sales Data Visualization"),

    # Dropdown to select store location
    dcc.Dropdown(
        id='store-location-dropdown',
        options=[{'label': location, 'value': location} for location in sales_data['store_location'].unique()],
        value='New York',  # Default value
        style={'width': '50%'}
    ),

    # Graph for visualizing sales over time
    dcc.Graph(id='sales-over-time'),

    # Graph for weather vs sales comparison
    dcc.Graph(id='weather-vs-sales')
])

# Callback to update sales over time graph based on selected location
@app.callback(
    Output('sales-over-time', 'figure'),
    Input('store-location-dropdown', 'value')
)
def update_sales_over_time(selected_location):
    filtered_data = sales_data[sales_data['store_location'] == selected_location]

    # Create a line plot of sales over time
    fig = px.line(filtered_data, x='date', y='sales_amount',
                  title=f'Sales Over Time for {selected_location}',
                  labels={'sales_amount': 'Sales Amount ($)', 'date': 'Date'})

    return fig

# Callback to update weather vs sales graph based on selected location
@app.callback(
    Output('weather-vs-sales', 'figure'),
    Input('store-location-dropdown', 'value')
)
def update_weather_vs_sales(selected_location):
    filtered_data = sales_data[sales_data['store_location'] == selected_location]

    # Create a scatter plot comparing sales amount and temperature
    fig = px.scatter(filtered_data, x='Temperature (°C)', y='sales_amount',
                     title=f'Weather vs Sales for {selected_location}',
                     labels={'sales_amount': 'Sales Amount ($)', 'Temperature (°C)': 'Temperature (°C)'})

    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
!pip install dash_bootstrap_components
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output
from pymongo import MongoClient

# MongoDB Connection
client = MongoClient("mongodb+srv://tsjannoun123:KufyyNNqnno0atX9@cluster0.sb8py.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0")
db = client["Sales_Weather_DAG"]
collection = db["Sales_Weather"]

# Query the data from MongoDB
data = list(collection.find())

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data)

# Ensure 'date' is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Check the first few rows of the DataFrame (Optional for debugging)
print(df.head())

# Ensure columns are correctly typed (important for the callback)
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
df['sales_amount'] = pd.to_numeric(df['sales_amount'], errors='coerce')

# Initialize the Dash app with Plotly.js included to avoid timeout issues
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# App layout
app.layout = html.Div([
    html.H1("Sales and Weather Data Analysis", style={'text-align': 'center'}),

    # Date picker range to filter the data
    dcc.DatePickerRange(
        id='date-picker-range',
        start_date=df['date'].min().date(),
        end_date=df['date'].max().date(),
        display_format='YYYY-MM-DD',
        style={'margin': '20px'}
    ),

    # Dropdown to select the store location
    dcc.Dropdown(
        id='store-location-dropdown',
        options=[{'label': location, 'value': location} for location in df['store_location'].unique()],
        value='New York',
        style={'margin': '20px'}
    ),

    # Graph showing sales and weather analysis
    dcc.Graph(id='sales-weather-graph'),

    # Additional analysis (e.g., Humidity vs Sales)
    dcc.Graph(id='humidity-sales-graph')
])

# Callback to update graph based on selected date range and store location
@app.callback(
    [Output('sales-weather-graph', 'figure'),
     Output('humidity-sales-graph', 'figure')],
    [Input('date-picker-range', 'start_date'),
     Input('date-picker-range', 'end_date'),
     Input('store-location-dropdown', 'value')]
)
def update_graph(start_date, end_date, selected_location):
    # Filter the data based on user inputs
    filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]  # Use 'date' column
    filtered_df = filtered_df[filtered_df['store_location'] == selected_location]

    # Create the first figure: Sales vs Temperature
    sales_weather_fig = px.scatter(filtered_df, x='Temperature', y='sales_amount', color='product_id',
                                   labels={'Temperature': 'Temperature (°C)', 'sales_amount': 'Sales Amount ($)'},
                                   title=f'Sales Amount vs Temperature for {selected_location}')

    # Create the second figure: Humidity vs Sales
    humidity_sales_fig = px.scatter(filtered_df, x='Humidity', y='sales_amount', color='product_id',
                                    labels={'Humidity': 'Humidity (%)', 'sales_amount': 'Sales Amount ($)'},
                                    title=f'Sales Amount vs Humidity for {selected_location}')

    return sales_weather_fig, humidity_sales_fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
